```python
import numpy as np
import matplotlib.pyplot as plt

def get_plot_from_percentatge(null_values, total_values):
    
    null_percentage = (null_values/total_values)*100
    not_null_percentage = 100- null_percentage
    
    list_values = [not_null_percentage, null_percentage]
    plt.hist(list_values, bins=2, edgecolor="black")
    
    plt.title("Tag Title")
    plt.xlabel("")
    
    

get_plot_from_percentatge(9, 13)
```
