```python
import pandas as pd
import numpy
import difflib
```

```python
df1 = pd.read_csv('10JS.csv')
df2 = pd.read_csv('10NoJS.csv')
df2['HTML'] = df2['HTML'].fillna('')
d = difflib.Differ()
```

```python
def get_difference(df1, df2):
    for i,id in enumerate (df1['idTool']):
        pos = df2[df2['idTool']==id].index.values[0]
        diff = d.compare(df1['HTML'].iloc[i], df2['HTML'].iloc[pos])
        html = ''
        difference = [l[2:] for l in diff if l.startswith('+ ') or l.startswith('- ')]
        html_difference = html.join(difference)
        yield html_difference
```

```python
htmls = [item for item in get_difference(df1, df2)]
htmls_df = pd.DataFrame(htmls, columns=['HTML'])
```
