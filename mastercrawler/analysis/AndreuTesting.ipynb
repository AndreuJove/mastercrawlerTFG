```python
def get_pie_chart(null_values, not_null_values):
    
    labels= "Null", "Not Null"
    sizes = [null_values, not_null_values]
    colors = ['#FF3300', 'yellowgreen']
    explode = (0, 0.1)

    
    plt.pie(sizes, explode=explode, labels=labels, colors=colors, labeldistance=1.11,
    autopct='%1.1f%%', shadow=True, startangle=130)

    plt.axis('equal')
    plt.show()
    
    
    
get_pie_chart(9, 4)
```
