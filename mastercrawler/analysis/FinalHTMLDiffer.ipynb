```python
import pandas as pd
import numpy
import difflib
```

```python
df1 = pd.read_csv('10JS.csv')
df2 = pd.read_csv('10NoJS.csv')
df2['HTML'] = df2['HTML'].fillna('')
d = difflib.Differ()
```

```python
def get_difference(df1, df2):
    for i,id in enumerate (df1['idTool']):
        pos = df2[df2['idTool']==id].index.values[0]
        diff = d.compare(df1['HTML'].iloc[i], df2['HTML'].iloc[pos])
        html_positive, html_negative = '',''
        
        for lletra in diff:
            if lletra.startswith('+ '):
                html_positive = html_positive + lletra[2]
                #len_html_positive = len(html_positive)
            elif lletra.startswith('- '):
                html_negative = html_negative + lletra[2]
        #positive_difference = [l[2:] for l in diff if l.startswith('+ ')]
        #negative_difference = [l[2:] for l in diff if l.startswith('- ')]

        #html_difference = html.join(difference)
        yield html_positive, html_negative, df1['idTool'].iloc[i]
```

```python
all_values = [item for item in get_difference(df1, df2)]
htmls_positive = [item[0] for item in all_values]
htmls_negative = [item[1] for item in all_values]
idTools = [item[2] for item in all_values]
```

```python
htmls_dictionary = {'HTML +' : htmls_positive, 'HTML - ' : htmls_negative, 'idTool' : df1['idTool']}
htmls_df = pd.DataFrame.from_dict(htmls_dictionary)
```
